```python
# TanH activation Function
def tanh(x):
    return np.tanh(x)
```

```python
x=600
print(tanh(x))
```

```python
x=np.linspace(-10000,10000,11)
print(x)
```

```python
p=tanh(x)
print(p)
```

```python
plt.xlabel('x')
plt.ylabel('tanh')
plt.plot(x,p)
plt.show()
```

maps input values to the range (-1,1)

```python
# ReLU
def relu(x):
    return np.maximum(0,x)
```

```python
x=10
relu(x)
```

```python
x=np.linspace(-10000,10000,11)
print(x)
```

```python
p=relu(x)
print(p)
```

```python
plt.xlabel('x')
plt.ylabel('relu')
plt.plot(x,p)
plt.show()
```

```python
# leaky_relu
def leaky_relu(x,alpha=0.01):
    return np.maximum( alpha*x,x)
```

```python
x=-6000
leaky_relu(x)
```

```python
x=np.linspace(-10000,10000,11)
print(x)
```

```python
p= leaky_relu(x)
print(p)
```

```python
plt.xlabel('x')
plt.ylabel('leaky_relu')
plt.plot(x,p)
plt.show()
```

```python
# Softmax
def softmax(x):
    exp_x=np.exp(x-np.max(x))
    return exp_x/ np.sum(exp_x, axis=0)
```

```python
x=-600
softmax(x)
```

```python
x=np.linspace(-10000,10000,11)
print(x)
```

```python
p=softmax(x)
print(p)
```

```python
x=np.array([2.0,1.0,0.5])
print(x)
```
